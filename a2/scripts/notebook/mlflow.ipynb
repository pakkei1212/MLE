```python
import mlflow
from mlflow.tracking import MlflowClient

mlflow.set_tracking_uri(os.getenv("MLFLOW_TRACKING_URI", "http://mlflow:5000"))
c = MlflowClient()

# List all registered models (2.x preferred)
try:
    regs = c.search_registered_models()  # MLflow 2.x
except AttributeError:
    regs = c.list_registered_models()    # fallback for very old versions

for rm in regs:
    # rm.latest_versions may be empty if no versions yet
    lv = getattr(rm, "latest_versions", []) or []
    print(f"- {rm.name}  (latest: {[f'v{v.version}:{v.current_stage}' for v in lv]})")
```
